```python
import requests
```

```python
#Getting html
base_url = "https://books.toscrape.com/catalogue/page-{}.html"
result = requests.get(base_url.format(1))
```

```python
#Parsing HTML
soup = bs4.BeautifulSoup(result.text,'lxml')
```

```python
import time
two_star_titles = []
for i in range(10):
    result = requests.get(base_url.format(i))
    soup = bs4.BeautifulSoup(result.text,'lxml')
    books = soup.select('.product_pod')
    for book in books:
        if len(book.select('.star-rating.Two'))!=0:
            two_star_titles.append(book.select('a')[1]['title'])
        
        
    
```

```python
price_list_three_star = []
for i in range(1,21):
    scrape_url = "https://books.toscrape.com/catalogue/page-{}.html"
    base_url = scrape_url.format(i)
    res = requests.get(base_url)
    soup = bs4.BeautifulSoup(res.text,'lxml')
    books = soup.select(".product_pod")
    for book in books:
        if len(book.select(".star-rating.Three")) != 0:
            for n in book.select('.product_price .price_color'):
                price_list_three_star.append((n.text,book.select('a')[1]['title']))
```
